```python
import pandas as pd
import numpy as np
from sklearn.datasets import make_regression
import seaborn as sns
import matplotlib.pyplot as plt
```

Genreating dataset

```python
X,y = make_regression(n_samples = 1000,n_features = 1,noise = 15)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2)
```

Scikit-Learn implementation of Lasso Regularisation 

```python
from sklearn.linear_model import Lasso
```

```python
m = Lasso(alpha = 0.1)
```

```python
m.fit(X_train,y_train)
```

```python
y_pred = m.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python
sns.scatterplot(x = X.reshape(-1,),y = y)
plt.plot(X_train,m.predict(X_train),c = 'black')
plt.title('Scikit-Learn Lasso Regularisation')
```

Mathematical Formulation of Lasso Regularisation 

```python
class lasso:
    def __init__(self,alpha):
        self.aplha = alpha
        self.coef_ = None
        self.intercept_ = None

    def fit(self,X_train,y_train):
        num = 0
        den = 0
        for i in range(X_train.shape[0]):
            num = num + ((X_train.iloc[i]-X_train.mean())*(y_train.iloc[i]-y_train.mean())-self.alpha)
            den = den + ((X_train.iloc[i]-X_train.mean())*(X_train.iloc[i]-X_train.mean()))
            self.coef_ = num/den
            self.intercept_ = y_train.mean()-self.coef_*X_train.mean()

    def predict(self,X_test):
        return self.coef_*X_test+self.intercept_
```

```python
ls = Lasso(alpha = 0.1)
```

```python
ls.fit(X_train,y_train)
```

```python
y_pred1 = ls.predict(X_test)
```

```python
r2_score(y_test,y_pred1)
```
