```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
from sklearn.datasets import make_regression
```

```python
X,y = make_regression(n_samples = 1000,n_features = 1,noise = 6)
```

```python
plt.scatter(X, y, marker='o', s=25)
plt.title("Generated Linear Data")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
array = np.concatenate((X,y.reshape(X.shape)),axis = 1)
```

```python
df = pd.DataFrame(array,columns = ['input','output'])
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df['input'],df['output'],test_size = 0.2)
```

Simple Linear Regression Mathematical Formulation 

```python
class Mylr:
    def __init__(self):
        self.m = None
        self.b = None

    def fit(self,X_train,y_train):
        num ,den = 0,0
        for i in range(X_train.shape[0]):
            num = num + ((X_train.iloc[i]-X_train.mean())*(y_train.iloc[i]-y_train.mean()))
            den = den + (X_train.iloc[i]-X_train.mean())**2
            self.m = num/den
            self.b = y_train.mean()-self.m*X_train.mean()

    def predict(self,X_test):
        return self.m*X_test + self.b
```

```python
lr = Mylr()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)                  # r2_score for own designed Mylr class
```

```python
from sklearn.linear_model import LinearRegression
```

```python
Lr = LinearRegression()
```

```python
Lr.fit(X_train.to_numpy().reshape(-1,1),y_train)
```

```python
y_pred1 = Lr.predict(X_test.to_numpy().reshape(-1,1))
```

```python
r2_score(y_test,y_pred1)                   #  r2_score for Scikit Learn Linear regression class
```
